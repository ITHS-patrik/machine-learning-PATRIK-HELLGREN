````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
samples = 1000

X = np.random.rand(samples, 2)
e = np.random.randn(samples)
y = 3*X[:, 0] + 5*X[:, 1] + 3 + e

plt.plot(X, y, '.')
plt.show()

X = np.c_[np.ones(len(X)), X]
````

## 1. Gradient descent - learning rate (*)

Use gradient descent to calculate $\vec{\theta} = (\theta_0, \theta_1, \theta_2)^T$ 

&nbsp; a) Use $\eta = 0.1$ and calculate $\vec{\theta}$ for each fifth epoch from 1 to 500. So the procedure is as follows:
- calculate $\vec{\theta}$ for epochs = 1
- calculate $\vec{\theta}$ for epochs = 6
- ...
- calculate $\vec{\theta}$ for epochs = 496

Plot these $\vec{\theta}$ values against epochs. (*)

&nbsp; b) Do the same as for a) but with learning rate $\eta = 0.01$, 5000 epochs and for each 20th epoch. What do you notice when changing the learning rate? (*)

&nbsp; c) Experiment with larger and smaller $\eta$ and see what happens.

````python
# A
y = y.reshape(-1, 1)

def gradient_descent(X, y, learning_rate=0.1, iterations=500, steps=5):
    n = len(X)
    theta = np.random.randn(X.shape[1],1)

    history = []
    epochs = []

    for epoch in range(1, iterations+1):
        gradient = 2 / n * X.T @ (X @ theta - y)
        theta -= learning_rate * gradient

        if epoch % steps == 0:
            history.append(theta.reshape(-1).copy())
            epochs.append(epoch)

    return theta, np.array(history), np.array(epochs)

theta, history, epochs = gradient_descent(X, y)
print(theta.reshape(-1))

plt.figure(figsize=(10,6))
plt.plot(epochs, history[:,0], label=r"$\theta_0$")
plt.plot(epochs, history[:,1], label=r"$\theta_1$")
plt.plot(epochs, history[:,2], label=r"$\theta_2$")

plt.xlabel("Epochs")
plt.ylabel("Theta value")
plt.title("Gradient descent - eta: 0.1")
plt.legend()
plt.grid(True)
plt.show()
````

````python
# B
theta, history, epochs = gradient_descent(X, y, learning_rate=0.01, iterations=5000, steps=20)
print(theta.reshape(-1))

plt.figure(figsize=(10,6))
plt.plot(epochs, history[:,0], label=r"$\theta_0$")
plt.plot(epochs, history[:,1], label=r"$\theta_1$")
plt.plot(epochs, history[:,2], label=r"$\theta_2$")

plt.xlabel("Epochs")
plt.ylabel("Theta value")
plt.title("Gradient descent - eta: 0.1")
plt.legend()
plt.grid(True)
plt.show()
````
